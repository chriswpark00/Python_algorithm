# column, index

```python
df = pd.DataFrame([
    ['dog', '멍멍이', '45kg'], 
    ['cat', '냥냥이', '7kg'], 
    ['bird', '짹짹이', '4g'],
    ['tiger', '큰냥이', '1t']
], index=['row1', 'row2', 'row3', 'row4'], columns=['종류', '이름', '무게'])
df
```

# loc, iloc
    - loc : 인덱스 명으로 접근
    - iloc : 인덱스 순번으로 접근

```python
df2.loc[0]
```

```python
df2.loc[[134, 32, 25]] #index이름으로 데이터 뽑기
```

```python
df2.iloc[0] #index번호로 데이터 뽑기
```

```python
df2.iloc[10:20]
```

```python
df.loc[['row1', 'row2']]
```

```python
df.iloc[0:3]
```
