# 특정 작품(titleId)의 특정회(no) 크롤링 하기

```python
url
```

```python
def download_naver_webtoon(titleId, no):
    url = f'https://comic.naver.com/webtoon/detail.nhn?titleId={titleId}&no={no}'
    response = requests.get(url)
    
    if response.status_code == 200:
        dom = BeautifulSoup(response.text, "html.parser")
        img = dom.select("div#comic_view_area .wt_viewer img")
        
        print(len(img), "개의 이미지")
        
        images = [
            i.attrs.get("src")
            for i in img
        ]
        
        dirName = 'download'
        
        for image in images:
            disassembled = urlparser.urlparse(image)
            filename = basename(disassembled.path)
            headers = {
                "Referer" : url,
                "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.129 Safari/537.36"
            }
            response_image = requests.get(image, headers = headers)
            
            print(response_image)
            
            dirName = f'download/{titleId}/{no}'
            if not os.path.exists(dirName):
                os.mkdir(dirName)

            if response_image.status_code == 200:
                with open(f"./{dirName}/{filename}", "wb") as f:
                    f.write(response_image.content)
        
```

```python
download_naver_webtoon(20853, 1236)
```

```python
for i in range(10):
    download_naver_webtoon(20853, 1226 + i)
```

```python
startNo = 1220
endNo = 1226
titleId = 20853

for no in range(startNo, endNo + 1):
    download_naver_webtoon(titleId, no)
```
